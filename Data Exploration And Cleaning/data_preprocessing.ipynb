We can create a function of what percentage of data is missing to create much easier for us to work.

```python
def percent_missing(df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90) #rotates the labels by 90 degrees.
```

### Step 2: Removal of least percentage missing data
Till now we have converted the percentage missing values and also visualized them.
Now, we will have to eliminate or update the missing data w.r.t. each corresponding column.


NOTE : Let's first work on considering feature with very small percentage missing. We can make some kind of threshold such as less than 1%.

```python
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
plt.ylim(0,1)
```

#### Trying to find out the Features with least missing data

```python
percent_nan[percent_nan < 1]
```

```python
df[df['Electrical'].isnull()]
```

```python
df = df.dropna(axis = 0, subset = ['Electrical', 'Garage Cars'])
```

```python
# calling percent_missing function to check what percentage of data is missing
percent_nan = percent_missing(df)
percent_nan[percent_nan < 1]
```

```python
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
plt.ylim(0,1)
```

NOTE : The plot displays the reduction of missing values

#### Try using domain Knowledge to explore around data

```python
df[df['Bsmt Half Bath'].isnull()]
```

```python
df[df['Bsmt Full Bath'].isnull()]
```

Cleaverly updating Bsmt(basement) values accordingly as this may be that the missing value indexes are the houses which do not have basement. Hence it's better to update the features accordingly.

```python
# BSMT NUMERIC COLUMNS --> 0
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Half Bath', 'Bsmt Full Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)

# BSMT STRING COLUMNS --> None
bsmt_str_cols = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')
```

```python
percent_nan = percent_missing(df)
percent_nan[percent_nan < 1]
```

```python
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
plt.ylim(0,1)
```

```python
df['Mas Vnr Type'] = df['Mas Vnr Type'].fillna('None')
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)
```

```python
percent_nan = percent_missing(df)
percent_nan[percent_nan < 1]
```

#### All of the features below the threshold of 1% are removed 

```python
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
```

### Step 3: Filling / Dropping feature values
* Dropping the feature column
* Filling in missing data : Statistical Estimation

```python
gar_str_cols = ['Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond']
```

```python
df[gar_str_cols] = df[gar_str_cols].fillna('None')
```

```python
# Upadating missing percentage and plotting
percent_nan = percent_missing(df)
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
```

```python
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(0)
```

```python
df = df.drop(['Pool QC', 'Misc Feature', 'Alley', 'Fence'], axis = 1)
```

```python
# Upadating missing percentage and plotting
percent_nan = percent_missing(df)
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
```

```python
df['Fireplace Qu'] = df['Fireplace Qu'].fillna('None')
```

```python
# Last feature left
df['Lot Frontage'] 
# Numeric Value
```

#### Using domain knowledge, we can group by:
* Neighborhood : Physical location
* Lot Frontage : Linear feet of street

```python
plt.figure(figsize = (8,12), dpi = 200)
sns.boxplot(x = 'Lot Frontage', y = 'Neighborhood', data = df, orient = 'h')
```

```python
df.groupby('Neighborhood')['Lot Frontage'].mean()
```

NOTE : Using transform method from pandas.DataFrame.Transform , no need to go through for loop to fill mean values. It makes it much faster and quicker.

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value : value.fillna(value.mean()))
```

```python
df.isnull().sum()
```

```python
# Some values still left , we can either drop the row or fill them with 0
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
```

```python
# Upadating missing percentage and plotting
percent_nan = percent_missing(df)
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90)
```

#### NOTE : If the above block throws the error of "ValueError: min() arg is an empty sequence" , it means that all the missing values have been updated successfully and we don't have anything left.

```python
df.to_csv("Ames_No_Missing_Data.csv")
```

### Step 4 : Categorical Data Conversion

```python
df = pd.read_csv('Ames_No_Missing_Data.csv')
```

#### We have to convert categorical features to Dummy variables and making it from dummy variables is much easier

```python
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

NOTE : In order to select each column and pass a list in direction area, we will use df select_dtype.

In python, string is considered as an object.

```python
df.select_dtypes(include = 'object')
```

```python
# string data
my_object_df = df.select_dtypes(include = 'object')

#numerical data
my_numerical_data = df.select_dtypes(exclude = 'object')
```

```python
df_object_dummies = pd.get_dummies(my_object_df, drop_first = True)
```

```python
df_object_dummies
```

#### Join back these results with numerical value

```python
final_df = pd.concat([my_numerical_data, df_object_dummies], axis = 1)
```

```python
final_df.corr()['SalePrice'].sort_values()
```
